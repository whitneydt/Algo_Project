```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decompose_close = seasonal_decompose(example_df.Close)
len(example_df)
# len(decompose_close) gives an Error because decompose_close has no length
```

```python
# RNN 
train = example_df.iloc[:348] # .7 * 497 equals 347.9 so I just rounded up since for the applied weight to the training
test = example_df.iloc[347:]
# Initialize for MinMaxScaler
scale = MinMaxScaler()
scale.fit(train)
train_scaled = scale.transform(train)
test_scaled = scale.transform(test)

length = 5
batch_size = 1
n_input = 1
n_features = 1 # Since we're using a univariate model

# After the generator object is made, initialize for Sequential & use the LSTM model
generator = TimeseriesGenerator(train_scaled, train_scaled, length=length, batch_size=batch_size)
X, y = generator[0]
X.shape # This is a 3D model
```

```python
# LSTM model
model = Sequential()
model.add(LSTM(100, activation='tanh', recurrent_activation='sigmoid', input_shape=(n_input, n_features)))
model.add(Dense(units=5))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
# Fit the model
model.fit(generator, epochs=30)
```
