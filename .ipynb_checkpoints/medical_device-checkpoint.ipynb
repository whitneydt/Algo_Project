```python
# Import libraries and dependencies
import numpy as np
import pandas as pd
import os
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import alpaca_trade_api as tradeapi
from pathlib import Path

%matplotlib inline
```

```python
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Set the random seed for resutls reproducibility (for testing purposes only)
np.random.seed(42)
```

```python
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()

# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Subset your tickers, then pull returns data:
# Set the ticker
ticker = ["MDT", "SYK"]

# Set timeframe to '1D'
timeframe = "1D"

# Set start and end datetimes of 1 year, between now and 365 days ago.
start_date = pd.Timestamp('2008-01-31', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-06-30', tz='America/New_York').isoformat()

# Get 1 year's worth of historical data for SKY and MDT
df = api.get_barset(
    ticker,
    timeframe,
    limit=None,
    start=start_date,
    end=end_date,
    after=None,
    until=None,
).df

# Drop unnecessary columns
df = df.drop(
    columns=['open', 'high', 'low', 'volume'],
    level=1)
df.head(20)
```

```python
# Calculate the daily roi for the stocks

daily_returns = df.pct_change()


mdt_daily_returns = daily_returns.pct_change()['MDT']['close']
syk_daily_returns = daily_returns.pct_change()['SYK']['close']

mdt_mean = daily_returns.mean()['MDT']['close']
syk_mean = daily_returns.mean()['SYK']['close']

# Display sample data
daily_returns.head(10)
```

```python
# Compute daily volatility
mdt_std_dev = daily_returns.std()['MDT']['close']
syk_std_dev = daily_returns.std()['SYK']['close']

print(f"Standard Deviation of MDT: {mdt_std_dev}")
print(f"Standard Deviation of SYK: {mdt_std_dev}")
```

```python
# Save the last day's closing price
last_mdt_price = df['MDT']['close'][-1]
last_syk_price = df['SYK']['close'][-1]
print(f"Last day's closing price for MDT: {last_mdt_price}")
print(f"Last day's closing price for SYK: {last_syk_price}")
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 100
number_records = 252 * 30
monte_carlo = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation
simulation_df = pd.DataFrame()

for n in range(number_simulations):
    
    simulated_mdt_prices = [last_mdt_price]
    simulated_syk_prices = [last_syk_price]
    
    for i in range(number_records):
        simulated_mdt_price = simulated_mdt_prices[-1] * (1 + np.random.normal(mdt_mean,mdt_std_dev))
        
        simulated_syk_price = simulated_syk_prices[-1] * (1 + np.random.normal(syk_mean,syk_std_dev))
        
        simulated_mdt_prices.append(simulated_mdt_price)
        simulated_syk_prices.append(simulated_syk_price)
                                                         
    simulation_df['mdt Prices'] = pd.Series(simulated_mdt_prices)
    simulation_df['syk Prices'] = pd.Series(simulated_syk_prices)
        
    simulated_daily_returns = simulation_df.pct_change()
        
    weights = [0.40, 0.60]
        
    portfolio_daily_returns = simulated_daily_returns.dot(weights)
        
    monte_carlo[n] = (1 + portfolio_daily_returns.fillna(0)).cumprod()
```

```python
# Check that the simulation ran successfully
monte_carlo.head()
```

```python
# Visualize the Simulation
monte_carlo.plot(legend = None, title = 'Monte Carlo Simulation')
```

```python
# Select the last row for the cumulative returns (cumulative returns at 30 years)
last_simulation = monte_carlo.iloc[-1,:]
last_simulation.head()
```

```python
# Display the 90% confidence interval for the ending returns
confidence_interval = last_simulation.quantile(q=[.05,.95])

plt.figure();
last_simulation.plot(kind='hist', density = True, bins = 10)
plt.axvline(confidence_interval.iloc[0], color = 'r')
plt.axvline(confidence_interval.iloc[1], color = 'r')
```
